```python
def find_minimum_vertex_cover(graph):
    n = len(graph)
    visited = [False] * n  # To track visited vertices
    cover_set = set()

    # Greedy approach: pick an edge and add both vertices to the cover
    for u in range(n):
        if not visited[u]:
            for v in range(n):
                if graph[u][v] == 1 and not visited[v]:  # If there's an edge and v is unvisited
                    cover_set.add(u + 1)  # Add u and v to the cover set (1-indexed)
                    cover_set.add(v + 1)
                    visited[u] = visited[v] = True  # Mark both as visited
                    break

    return cover_set

def run_test(graph):
    # Find and print the result
    vertex_cover = find_minimum_vertex_cover(graph)

    if vertex_cover:
        print(" ".join(map(str, vertex_cover)))  # Print the vertices in the cover set
        print(len(vertex_cover))  # Print the size of the cover set
```
